```python
!pip install opencv-python
import cv2
from google.colab.patches import cv2_imshow
from ultralytics import YOLO
import numpy as np
'''
from google.colab import files

import cv2
import numpy as np
from google.colab.patches import cv2_imshow


# Upload model and image
uploaded = files.upload()

# Load model and image
model_path = 'last.pt'  # Update this if the uploaded file has a different name
image_path = 'wolf.jpg'  # Update this if the uploaded file has a different name
'''
img_name = 'inf2'
img_path = '/content/gdrive/My Drive/yolov8_pose/inference/'+img_name+'.jpg'
model = YOLO('/content/gdrive/My Drive/yolov8_pose/inference/best_150000.pt')

img_cv = cv2.imread(img_path)


# Run the model
results = model(img_path)

corner_xy_list = []

for result in results:
  corner_xy_list.append(result.keypoints.xy)

corner_xy_list = np.asarray(corner_xy_list[0][0])


# Annotate image with keypoints
for keypoint_indx, keypoint in enumerate(corner_xy_list):
    cv2.putText(img_cv, str(keypoint_indx), (int(keypoint[0]), int(keypoint[1])),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)


# Display the result
cv2_imshow(img_cv)

```

```python
img_out_name = img_name+'_output.jpg'
cv2.imwrite(img_out_name, img_cv)

import locale
locale.getpreferredencoding = lambda: "UTF-8"

```
